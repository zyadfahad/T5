```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose


import statsmodels.api as sm
```

## Read dataset

```python
traffic = pd.read_csv('/content/traffic.csv', index_col ='DateTime',parse_dates = True)
```

## show dataset

```python
traffic.head()
```

## Apply ETS

```python
result = seasonal_decompose(
                           traffic['Vehicles'],
                            model ='multiplicative',period = 128)
```

## Plot ETS

```python
# prompt: plot the ets

result.plot()
plt.show()
```
